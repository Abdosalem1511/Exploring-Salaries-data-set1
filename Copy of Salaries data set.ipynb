```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the salaries dataset
salaries_df = pd.read_csv('Salaries.csv')
```

```python

# Basic Data Exploration
num_rows = salaries_df.shape[0]
num_cols = salaries_df.shape[1]
data_types = salaries_df.dtypes
missing_values = salaries_df.isnull().sum()
```

```python

# Descriptive Statistics
mean_salary = salaries_df['TotalPay'].mean()
median_salary = salaries_df['TotalPay'].median()
mode_salary = salaries_df['TotalPay'].mode().values[0]
min_salary = salaries_df['TotalPay'].min()
max_salary = salaries_df['TotalPay'].max()
salary_range = max_salary - min_salary
salary_std = salaries_df['TotalPay'].std()
```

```python

# Basic Data Visualization
plt.hist(salaries_df['TotalPay'], bins=20)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries')
plt.show()

department_counts = salaries_df['JobTitle'].value_counts().head(10)
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python

# Grouped Analysis
grouped_data = salaries_df.groupby('Year').mean()
average_salaries_by_year = grouped_data['TotalPay']
```

```python

# Simple Correlation Analysis
correlation = salaries_df['TotalPay'].corr(salaries_df['BasePay'])
plt.scatter(salaries_df['BasePay'], salaries_df['TotalPay'])
plt.xlabel('Base Pay')
plt.ylabel('Salary')
plt.title('Correlation between Base Pay and Salary')
plt.show()
```
